## 3 - Declare Variable

In this section, I work on just number categorical variables and I try to clean and preprocess them. The number categorical variables of this dataset are:</br>
<li> Signal Strength

```python
CatNum_list = ['Signal Strength']
```

## 4 - Data Type Conversion

the first thing that should be check is that we must make sure about data type. Just because the column show numbers, it doesn't mean that they are integer.

```python
def non_numeric(x):
    non_numeric_df = pd.DataFrame(df[df[x].astype(str).str.contains('[^0-9]+')])
    return non_numeric_df
```

run the function for this variable:

```python
non_numeric('Signal Strength')
```

in this section, we don't have work to do anything in null value. let's clean some values that contain string.

```python
df['Signal Strength'] = df['Signal Strength'].str.replace('[^0-9]', '')
```

## 5 - Handle Null Values

we must make sure whether we have null values in these variables or not.

```python
def null_checker():
    for i in range(len(CatNum_list)):
        print(CatNum_list[i] , ':' , df.loc[df[CatNum_list[i]].isnull()].index)
```

Then we run the function for variables

```python
null_checker()
```

the result shows we have null values.first we should make sure whether the number of missing values of each variable, are less then 5% of whole data or more

```python
for i in range(len(CatNum_list)):
    count_nan = df[CatNum_list[i]].isna().sum()
    print(CatNum_list[i] , ':' , round((count_nan / df.shape[0]) * 100, 2) , '%')
```

### 5.1 - Drop Null

the results show, we can remove the rows, because they are less than 5%.

```python
print("the dimension before removing null: " , df.shape)
```

now we remove null values:

```python
df = df.dropna(subset=['Signal Strength'])
df = df.reset_index(drop=True)
print("the dimension after removing null: " , df.shape)
```

now again we can check the statuse of null values:

```python
null_checker()
```

finally, we convert the data type

```python
df['Signal Strength'] = df['Signal Strength'].astype(int)
```

## 6 - Possible Range

the "Signal Strength" should be between 0 to 5, based on data documentation.

```python
print('Range of Height', sorted(df['Signal Strength'].unique()))
print('\n')
print(df['Signal Strength'].value_counts())
```

since, this variable means categorical in nature, we should find the mode, and impute our-range values.

```python
df['Signal Strength'] = df['Signal Strength'].map(lambda x: x if 0 <= x <= 5 else 4)
print('Range of Height', sorted(df['Signal Strength'].unique()))
```
