## 4 - Data Type Conversion

```python
def non_numeric(x):
    non_numeric_df = pd.DataFrame(df[df[x].astype(str).str.contains('[^0-9]+')])
    return non_numeric_df
```

### 4.5- Implement Conversion

now we should convert this variable to integer.

```python
def convert_to_number(x):
    #conversion
    df[x] = df[x].fillna(0).replace(0, np.nan).astype(float)
    #return the sum of the column
    return df[x].sum()

for_range = len(numerical_list) 
for i in range(for_range):
    x = df[numerical_list[i]].name
    convert_to_number(x)
    
    print(numerical_list[i] , ':' , df[numerical_list[i]].sum())
```

## 5 - Handle Null Values

First we must make sure whether we have null values in these variables or not.

```python
def null_checker():
    for i in range(len(numerical_list)):
        print(numerical_list[i] , ':' , df.loc[df[numerical_list[i]].isnull()].index)
```

Then we run the function for all numerical variables

```python
null_checker()
```

the result shows we have null values in duration, cost and team member.first we should make sure whether the number of missing values of each variable, are less then 5% of whole data or more.

```python
for i in range(len(numerical_list)):
    count_nan = df[numerical_list[i]].isna().sum()
    print(numerical_list[i] , ':' , round((count_nan / df.shape[0]) * 100, 2) , '%')
```

### 5.1 - Drop Null

the results show, about Duration and team Member we can remove the rows, because they are less than 5%.

```python
print("the dimension before removing null: " , df.shape)
```

now we remove null values for duration and team member:

```python
df = df.dropna(subset=['Duration' , 'Team Member'])
print("the dimension after removing null: " , df.shape)
```

now again we can check the statuse of null values:

```python
null_checker()
```

### 5.2 - Impute Null

Now, we should behave with Cost, since the number of its null values is more than 5%. for doing this, we can impute the null value with mean or median. Let's see its skenwss:

```python
print('Skewness :' , round(df['Cost'].skew() ,3))

mean_cost = df['Cost'].mean()
median_cost = df['Cost'].median()

if mean_cost > median_cost:
    print('Mean is bigger than Median. Left Skewed. Median for imputing')
else:
    print('Mean is smaller than Median. Right Skewed. Mean for imputing')
```

```python
df['Cost'] = df['Cost'].fillna(mean_cost).round(0)
```

now check the number of null values:

```python
null_checker()
```
