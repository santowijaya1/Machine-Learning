## 3 - Declare String Categorical Variables

we should declare all categorical variables that need to be mapped to number. so, we list lable categorical variable:
*   Frequency (ordinal)
*   Antenna Type (nominal)
*   Orientation (nominal)
*   Power Supply (nominal)
*   Zone (nominal)

## 4 - Map variables to number

### 4.1 - Frequency

This variable is ordinal, so it means we should consider the order of the categories.

```python
df['Frequency'] = df['Frequency'].map(
    {'Very Low Frequencies (VLF)':0, 
    'Low Frequencies (LF)' : 1,
    'Medium Frequencies (MF)' : 2,
    'High Frequencies (HF)' : 3,
    'Very High Frequencies (VHF)' : 4,
    'Ultra-High Frequencies (UHF)' : 5,
    'Super-High Frequencies (SHF)' : 6,
    'Extremely High Frequencies (EHF)' : 7}
    )
```

### 4.2 - Antenna Type

This variable is nominal, so it means there is no difference among the orders of categories.

```python
df['Antenna Type'] = df['Antenna Type'].map(
    {'Dielectric':0, 
    'Printed Circuit Board (PCB)' : 1,
    'Wire' : 2,
    'Aperture' : 3,
    'Reflector' : 4,
    'Array' : 5}
    )
```

### 4.3 - Orientation

This variable is nominal, so it means there is no difference among the orders of categories.

```python
df['Orientation'] = df['Orientation'].map(
    {'Omni-directional':0, 
    'Circular' : 1,
    'Horizontal' : 2,
    'Vertical' : 3}
    )
```

### 4.4 - Power Supply

This variable is nominal, so it means there is no difference among the orders of categories.

```python
df['Power Supply'] = df['Power Supply'].map(
    {'Solar-powered':0, 
    'Active' : 1,
    'Battery-powered' : 2,
    'PoE' : 3,
    'Passive': 4}
    )
```

### 4.5 - Zone

This variable is nominal, so it means there is no difference among the orders of categories.

```python
df['Zone'] = df['Zone'].map(
    {'North':0, 
    'Center' : 1,
    'West' : 2,
    'South' : 3,
    'East': 4}
    )
```
