## 3 - Check Multicollinearity

After analysin the correlation between dependent variable (cost) and other independent variables, now we should check correlation between all independent variables. It means we must check the multicollinearity. For doing this, we use heatmap plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (18,5))
sns.heatmap(df.corr(),
            vmin = -1,
            vmax = 1,
            cmap = "GnBu",
            annot = True)
plt.show()
```
